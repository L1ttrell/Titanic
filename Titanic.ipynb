```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression as LOR
from xgboost import XGBClassifier as xgb
from sklearn.metrics import accuracy_score
from sklearn.grid_search import GridSearchCV
```

```python
data_train = pd.read_csv('D:/kaggle/datas/Titanic/train.csv')
data_test = pd.read_csv('D:/kaggle/datas/Titanic/test.csv')
data_test_svv = pd.read_csv('D:/kaggle/datas/Titanic/gender_submission.csv')
data_test = pd.merge(data_test, data_test_svv)
```

```python
data_train['Gender']=data_train['Sex'].map({'male':1, 'female':0})
data_train['Aboard']=data_train['Embarked'].map({'S':0, 'C':1, 'Q':2})
data_train = data_train[['Survived', 'Gender', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Aboard']]

data_test['Gender']=data_test['Sex'].map({'male':1, 'female':0})
data_test['Aboard']=data_test['Embarked'].map({'S':0, 'C':1, 'Q':2})
data_test = data_test[['Survived', 'Gender', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Aboard']]
```

```python
model = xgb(n_estimators=100, max_depth=3, alpha=1, learning_rate=0.1)
```

```python
model.kwargs
```

```python
model.fit(data_train.drop('Survived', axis=1), data_train['Survived'])
```

```python
accuracy_score(model.predict(data_train.drop('Survived', axis=1)), data_train['Survived'])
```

```python
y_pred = model.predict(data_test.drop('Survived', axis=1))
accuracy_score(y_pred, data_test['Survived'])
```
